```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('../log.txt', engine='python', skiprows=1, skipfooter=4, header=None)
df.columns = ['it', 'loss', 'time']

df['it'] = df['it'].str[4:].astype(int)
df['loss'] = df['loss'].str[6:].astype(float)
df['time'] = df['time'].str[6:].astype(float)

df.head()
```

```python
df_ = df[df['it'] <= 8000]
plt.plot(df_['it'], df_['loss'])
plt.yscale('log')
plt.gcf().set_size_inches(20,10)
```
